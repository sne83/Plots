```python
# Univariate plot
import matplotlib.pyplot as plt
# pyplot is submodule for 2D plottings
```

```python
plt.plot([20,60,10,150])
```

```python
# groupby Cylinders
data = df.groupby('Cylinders')['CO2 Emissions(g/km)'].mean()
data
```

```python
plt.plot(data,marker='*',color='red',linestyle='--',label='Co2 Emission/cyl')
#plt.text(3,180,'Pallavi')
for x,y in zip(data.index,data.values):

    label = "{:.2f}".format(y)

    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center')
plt.legend()
```

```python
plt.plot(data,marker='*',color='red',linestyle='--',label='Co2 Emission/cyl')
plt.title('Co2 Emission per Cylinder')
plt.xlabel('Cylinders')
plt.ylabel('CO2 Emission')
plt.grid()
plt.legend()

# to save the plot
plt.savefig('p.jpg')
```
