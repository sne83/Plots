# Seaborn
* open source, BSD-licensed python library providing high level API for visualizing the data using python programming language.
* It is built on top of matplotlib

**Seaborn vs matplotlib:**
* It is summarized that if matplotlib 'tries to make easy things easy & hard things possible', Seaborn tries to make a well-defined set of hard things easy too.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# how to know dataset names
print(sns.get_dataset_names())
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.species.value_counts()
```

```python
# load diamond dataset
sns.load_dataset('diamonds').head()
```

**Plotting using seaborn**

```python
df.columns
```

```python
# flexibly plot a univariate distribution of observation
sns.distplot(df.sepal_length) # combo of histo & density    # old
```

```python
sns.displot(df.sepal_length)   # use this plot not above one
```

```python
sns.displt(df.sepal_length, bins=[])
```

```python
# sns.displot(df.petal_length)   # direct approach
sns.displot(data=df, x='petal_length')   # separate approach ----> go with this
```

**Conclusion:**
* between 2 to 3 we hv missing data
* above three it follows normal distribution
* range 4 to5.5 has more records
* min 1 & max 7
* max record we have in range>1 and <2

```python
# hist plot on all data
a = sns.displot(data=df, palette = 'PuBu')          # can also check for df.sepal_length
a.set_titles('Iris data hist plot')
a.fig.suptitle('Iris dataset')
a.set_xlabels('Features')
a.set_ylabels('Feature count',fontsize=12)
```

```python
sns.displot(data=df['petal_length','sepal_length'])
```

```python
# KDE = kernel density - helps to indentify the population (frequency distribution)
sns.kdeplot(df.petal_width)
```

```python
sns.kdeplot(data=df)  # for all cols
```

```python
#
sns.histplot(data=df.sepal_length)
```

Joint

```python
# it gives scattre plot & its strictly bivariate
sns.jointplot(x='petal_length',y='petal_width',data=df)
```

kde = kernel density estimation is a non-parametric way to estimate the distribution of a variable

```python
t = sns.load_dataset('tips')
t.head()
```

```python
sns.jointplot(x='total_bill',data=t,kind='kde')
```

```python
# daywise count
sns.countplot(data=t, x='day',palette='Dark2',hue='sex')
```
